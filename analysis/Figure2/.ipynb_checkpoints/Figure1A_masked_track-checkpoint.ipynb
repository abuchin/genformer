```python
def one_hot(sequence):
    '''
    convert input string tensor to one hot encoded
    will replace all N character with 0 0 0 0
    '''
    vocabulary = tf.constant(['A', 'C', 'G', 'T'])
    mapping = tf.constant([0, 1, 2, 3])

    init = tf.lookup.KeyValueTensorInitializer(keys=vocabulary,
                                               values=mapping)
    table = tf.lookup.StaticHashTable(init, default_value=0)

    input_characters = tfs.upper(tfs.unicode_split(sequence, 'UTF-8'))

    out = tf.one_hot(table.lookup(input_characters),
                      depth = 4,
                      dtype=tf.float32)
    return out

one_hot_encode('T').numpy()
```

```python
def removeElement(nums, val):
    """
    :type nums: List[int]
    :type val: int
    :rtype: int
    """
    i = 0
    half = (len(nums) - 1) //2
    end = len(nums) - 1
    while (end >= half and i <= half):
        if (nums[i] == val and nums[end] != val):
            nums[i] = nums[end]
            nums[end] = val
            end -= 1
            i += 1
        elif (nums[i] == val and nums[end] == val):
            while (end >=0 and nums[end] == val):
                end -= 1
            nums[i] = nums[end]
            nums[end] = val
            i += 1
        else:
            i += 1
            continue
        
    return end
            

lst = [3,2,2,3]
num_vals = removeElement(lst, 2)
print(num_vals)
print(lst[:num_vals].sort())
```
